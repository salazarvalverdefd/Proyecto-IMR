```python
numeros_serie = []
for j in data['train']['X_train']:
  x = j.split(".")
  #print(x)
  numeros_serie.append(x[5])

print("Numero de serie:",numeros_serie)
```

```python
def unico(x,L):
  esUnico = True
  for i in range(len(L)):
    if x == L[i]:
      esUnico = False
      break
  return esUnico
```

```python
def read_and_write(in_dir,read_video,write_video):
  path_origen = in_dir + read_video
  path_destino = in_dir + write_video
  shutil.copy(path_origen, path_destino)
```

```python
# Moviendo los archivos
def mov_files(X,in_dir,end_dir):
  for i in X:
    shutil.move(in_dir + i,end_dir)
  print("Exito!.Los archivos se movieron al directorio:",end_dir)
```

```python
def balancear(X,y,clase,multiplicador,add,dir):
  
  # Multiplicador
  for i in tqdm(range(0,len(X))):
    split_name = X[i].split(".")
    tipo_registro = split_name[0].split("_")
    if tipo_registro[0] == clase:
      for j in range(0,multiplicador):
        k = 0
        while k < 1:
          x = random.randint(10000,99999)
          if unico(x,numeros_serie):
            numeros_serie.append(x)
            new_name_vid = '.'.join(split_name[0:5]) + str(".") + str(x) + ".mp4"
            X.append(new_name_vid)
            y.append(0)
            #read_and_write(in_dir,X[i])
            #read_and_write(in_dir,'.'.join(split_name[0:5]) + str(".") + str(x) + ".mp4")
            shutil.copy(dir + X[i], dir + new_name_vid)
            k += 1
  
  # Aleatoridad
  c = list(zip(X, y))
  random.shuffle(c)
  X, y = zip(*c)
  X = list(X)
  y = list(y)

  # Adicional
  n = 0
  for i in range(0,len(X)):
    while n<5:
      split_name = X[i].split(".")
      tipo_registro = split_name[0].split("_")
      if tipo_registro[0] == clase:
        p = 0
        while p < 1:
            x = random.randint(10000,99999)
            if unico(x,numeros_serie):
              numeros_serie.append(x)
              new_name_vid = '.'.join(split_name[0:5]) + str(".") + str(x) + ".mp4"
              X.append(new_name_vid)
              y.append(0)
              shutil.copy(dir + X[i], dir + new_name_vid)
              p += 1
              n += 1
      break
  print("nnn:",n)
  return X,y
```
